```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('IMDB-movies.csv')
```

```python
data
```

```python
data=data.drop(['original_language','popularity','release_date','vote_average','vote_count'],axis=1)
```

```python
data
```

```python
data['tags']=data['overview']+data['genre']
```

```python
data
```

```python
data=data.drop(['genre','overview'],axis=1)
```

```python
data
```

```python
cv=CountVectorizer(max_features=1000,stop_words='english')
```

```python
vector=cv.fit_transform(data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similarity_matrix=cosine_similarity(vector)
```

```python
similarity_matrix
```

```python
#for example Captain America
d= sorted(list(enumerate(similarity_matrix[9997])), reverse=True, key= lambda vector: vector[1])
for i in d[1:6]:
  print(data.iloc[i[0]].title)
```

```python
def recommend(movie):
  i=data[data['title']==movie].index[0]
  x=sorted(list(enumerate(similarity_matrix[i])), reverse=True, key=lambda vector: vector[1])
  for j in x[1:6]:
    print(data.iloc[j[0]].title)
```

```python
recommend('The Godfather: Part II')
```

```python
recommend('The Conjuring')
```

```python
import pickle
pickle.dump(data, open('movies_list.pkl', 'wb'))
pickle.dump(similarity_matrix,open('similarity.pkl','wb'))
pickle.load(open('movies_list.pkl','rb'))
```
